```python
import numpy as np
import matplotlib.pyplot as plt
```

# Exercise 1

```python
#Brownianmotion function
def brownianmotion(X, delta): # X=(x,y)
    liste_positions = [X]
    i = 0
    while ((np.sum(liste_positions[i]**2) <=0.3) & (i <1000)):
        nouvelle_position = X + np.sqrt(delta)*np.random.randn(2)
        liste_positions.append(nouvelle_position)
        i+=1
    return (liste_positions)

a = brownianmotion(np.zeros(2),0.01)
array = np.array(a)
plt.plot(a)
print(len(a))
print(array[:][0])
    
    
```
